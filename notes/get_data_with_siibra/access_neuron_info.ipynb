# Cortical cell body distributions

```python
# Importing the necessary libraries
import siibra
import matplotlib.pyplot as plt
from nilearn import plotting
```

```python
# Getting the region of interest
v1 = siibra.get_region("julich 3.1", "v1")

# Getting the cell density profile of the region
cf = siibra.features.get(v1, siibra.features.cellular.CellDensityProfile)[0]
print(cf.name)
```

```python
# Listing attributes and methods of the cell density profiles
dir(cf)
```

```python
# Listing attributes and methods of one of the cell density profiles
dir(cf[0])
```

```python
print(cf[0].description)
```

```python
# Segmented cell body density mean (cells / 0.1mm3)
print(cf.data)

# Segmented cell body density (cells / 0.1mm3)
print(cf[0].data)
```

```python
# Plotting the cell density profile
cf[1].plot()
```

```python
cells = cf[1].cells
print("Number of segmented cells:", len(cells))
cells.head()
```

```python
# Plotting the cell distribution
plt.scatter(cells.x, cells.y, c=cells.layer, s=0.2)
plt.title(f"Cell distributions in {v1.name}")
plt.grid(True)
plt.axis("equal")
plt.tight_layout()
```

```python
# Getting information about the layers: Area and thickness
cf[1].layers
```

```python
# Cells in each layer
grouped_cells = cells.groupby('layer')
grouped_cells.size()
```

```python
# Volume per layer in microns^3
vol_per_layer = cf[1].layers
vol_per_layer['Vol(micron**3)'] = cf[1].layers["Area(micron**2)"] * cf[1].layers["AvgThickness(micron)"]
print(vol_per_layer)
```
